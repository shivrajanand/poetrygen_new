```python
# 3 patterns 
# meter (
# meter [
# meter : 

# Lets search them and see if there are other pattern than these 
import re
pattern = r'^.+?[\(\[:]'


f = open('skr_out_without_bracket.txt', 'a')

for skr_out in df['skr_out'].unique():
    if bool(re.match(pattern, skr_out)):
        continue
    f.write(skr_out+"\n")
    
# Empty file result that means skrutable result only in these 3 formats
```

```python
# Extracting meter name from these patterns 

# Two more patterns found; meter = and meter +

pattern = re.compile(r'^(.*?)\s*[\(\[:=+]')

def extract_meter(text):
    if not isinstance(text, str):
        return None

    m = pattern.match(text)
    if m:
        return m.group(1).strip()
    return None
```

```python
df["skr_meter"] = df["skr_out"].apply(extract_meter)
```

```python
import sys
from pathlib import Path

sys.path.append(str(Path("chandas-detector").resolve()))

import chandas_detector
```

```python
from chandas_detector import detect_meter, format_result

verse = """
काष्ठं जनः कुलधनैरनुरञ्जनीयस्
तन्नो यदुक्तमशिवं न हि तत्क्षमं ते।
नैसर्गिकी सुरभिणः कुसुमस्य सिद्धा
मूर्ध्नि स्थितिर्न चरणैरवताडनानि
"""

result = detect_meter(verse)

if result.confidence == "exact":
    print(f"Detected: {result.meter}")
else:
    print("Could not verify meter.")

# Print the formatted text diagnostic representation
print(format_result(result))
```

```python
result.meter
```

```python
meter_codes = []

for r in tqdm(df.itertuples(index=False), total=len(df)):
    verse = r.sa.strip()
    result = detect_meter(verse)

    meter_codes.append(result.meter if result.confidence == "exact" else None)

df["meter_cd"] = meter_codes
```

```python
valid_syllable_count = {
    'Vasantatilakā':56,
    'Anuṣṭubh':32,
    'Śārdūlavikrīḍita':76,
    'Mālinī':60,
    'Indravajrā':44,
    'Śālinī':44,
    'Vaṃśastha':48,
    'Śikhariṇī':68,
    'Upendravajrā':44,
    'Sragdharā':84
}
```

```python
df.isnull().sum()
```

```python
df["comments"] = None

for row in tqdm(df.itertuples(index=True), total=len(df)):
    comments = []

    if row.meter_cd is not None:
        if (
            row.skr_meter is None
            or row.meter_cd.strip().lower() != row.skr_meter.strip().lower()
        ):
            comments.append("Skrutable Mismatch")

        if row.syllable_count != valid_syllable_count[row.meter_cd]:
            comments.append("Syllable Count Error")

    if comments:
        df.at[row.Index, "comments"] = " + ".join(comments)
```

```python
import pandas as pd
```

```python
examples = (
    df[df["split"] == "train"]
    .groupby("meter_cd", group_keys=False)
    .head(3)
)

examples
```

```python
df = df.drop(examples.index).reset_index(drop=True)
```

```python
print(df.loc[0]['prompt'])
```

```python
example_map = {}

for meter, group in examples.groupby("meter_cd"):
    prompt_examples = ""

    for i, (_, row) in enumerate(group.iterrows(), start=1):
        prompt_examples += f"""
Example {i}:
Meaning:
{row['hi']}

Sanskrit Verse:
{row['sa']}

"""

    example_map[meter] = prompt_examples
```

```python
print(example_map['Anuṣṭubh'])
```

```python
def update_prompt(row):
    return row["prompt"] + example_map[row["meter_cd"]]

df["prompt"] = df.apply(update_prompt, axis=1)
```
