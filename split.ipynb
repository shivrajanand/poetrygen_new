```python
import json

with open("chanda_rules.json", "r", encoding="utf-8") as f:
    meter_rules = json.load(f)
```

```python
meter_rules.keys()
```

```python
df.isnull().sum()
```

```python
def build_prompt(target_language, source_language, chandas, rules):
    prompt = f"""Objective:

Generate a {target_language} verse in {chandas} meter based on a given meaning or theme while strictly following the metrical (?and rhyme) rules.

Rules of Chandas:
{rules}

Language:
The generated verse must be in {target_language} in Devanagri script.
The Input given is in the {source_language}.
Do NOT include English words, transliteration, labels, or explanations.

Output Format:
(Output exactly 4 lines, one pada per line. Nothing else.)
"""
    return prompt
```

```python
row = df.iloc[0]

row
```

```python
print(build_prompt('sanskrit', 'hindi', row['meter_cd'], meter_rules[row['meter_cd']]))
```

```python
df["prompt"] = df.apply(
    lambda row: build_prompt(
        "sanskrit",
        "hindi",
        row["meter_cd"],
        meter_rules[row["meter_cd"]]
    ),
    axis=1
)
```

```python
df['prompt'].nunique()
```

```python
df.to_csv("Files/v3_gitapress_final.csv", index=False)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split

df = pd.read_csv("Files/v3_gitapress_final.csv")

# First split: 80% train, 20% temp
train_df, temp_df = train_test_split(
    df,
    test_size=0.2,
    stratify=df["meter_cd"],
    random_state=42,
)

# Second split: split the remaining 20% into 10% val and 10% test
val_df, test_df = train_test_split(
    temp_df,
    test_size=0.5,
    stratify=temp_df["meter_cd"],
    random_state=42,
)

print(len(train_df), len(val_df), len(test_df))
```

```python
df["split"] = ""

df.loc[train_df.index, "split"] = "train"
df.loc[val_df.index, "split"] = "val"
df.loc[test_df.index, "split"] = "test"
```

```python
df.to_csv("Files/v3_gitapress_final.csv", index=False)
```
